```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
print(df.head())
print(df.isnull().sum())
```

```python
# df = df.dropna(subset=['Age'])
df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df)
```

```python
survived = df[df["Survived"] == 1]["Age"]
print(survived)
died = df[df["Survived"] == 0]["Age"]
print(died)
```

```python
plt.figure(figsize=(10, 4))
plt.hist(survived, bins=20, alpha=0.6, label="Survived", color="green")
plt.hist(died, bins=20, alpha=0.6, label="Died", color="red")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Survival by Age — Titanic")
plt.legend()
plt.show()
```

```python
print("Average age survived:", np.mean(survived.values))
print("Average age died:", np.mean(died.values))
```
