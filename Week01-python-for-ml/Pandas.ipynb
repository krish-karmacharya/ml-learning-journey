Series
A pandas 1-dimesnional labeled array that can hold any data type 

loc = location by label, iloc = locate by integer postion , length of data and index must be same

```python
data =[100.1,102,104,100]
series = pd.Series(data, index=['a','b','c','d'])
series.loc["a"] = 200
print(series.loc['a'])
```

```python
print(series.iloc[1])
print(series[series >= 102])
print(series[series < 102])
```

```python
calories ={"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series = pd.Series(calories)
series.loc["Day 3"] += 300
series.iloc[1] += 100
print(series)
print(" /n")
print(series[series> 1800])
```
